```python
# Generate a bar plot showing the total number of measurements taken 
# on each drug regimen using pandas.

# this gets the absolute total measurements per each regimine,
# one for each and every data measurement
def get_absolute_total_measurements_per_regimen(data):
    drug_reg = 'Drug Regimen'
    df = pd.DataFrame()
    for name, group in mouse_study_data.groupby(drug_reg):
        measurements = 0
        # add 5 measurements per unique mouse
        # one for Mouse ID, Drug Regimen, Sex, Age_months, and Weight (g)
        measurements += group['Mouse ID'].nunique() * 5
        # add the non-null values for Timepoint
        measurements += group['Timepoint'].count()
        # add the non-null values for Tumor Volume (mm3)
        measurements += group['Tumor Volume (mm3)'].count()
        # add the non-null values for Metastatic Sites
        measurements += group['Metastatic Sites'].count()
        total_df = pd.DataFrame({drug_reg: [name],
                              'Total Measurements Taken': measurements})
        df = pd.concat([df, total_df])
    return df.set_index(drug_reg)

# get the measurements for each mouse, using the 'Timepoint' as the way to count
total_measurements_per_regimen = mouse_study_data.groupby('Drug Regimen')['Timepoint'].count()
total_measurements_per_regimen
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint

# Get the data in the dataframe that are the final timepoints
def get_final_timepoints(data):
    tp = 'Timepoint'
    finals = pd.DataFrame()
    # group the data by mouse and drug
    for (drug, mouse), group in data.groupby(['Drug Regimen','Mouse ID']):
        # use the max timepoint value to get the whole row
        max_timepoint_row = group.loc[group[tp] == max(group[tp])]
        # add to finals
        finals = pd.concat([finals, max_timepoint_row])
    return finals

# get the final timepoints, then filter down 
# to 'Drug Regimen' and 'Tumor Volume (mm3)'
tumor_vol_by_drug = get_final_timepoints(mouse_study_data)[['Drug Regimen','Tumor Volume (mm3)']]
# set the index to 'Drug Regimen'
tumor_vol_by_drug.set_index('Drug Regimen', inplace=True)
# rename for clarity
tumor_vol_by_drug.rename(columns={'Tumor Volume (mm3)': "Final Tumor Volume (mm3)"}, inplace=True)
# display sumary just to see data
tumor_vol_by_drug.groupby('Drug Regimen').describe()
```

```python
# Create boxplot to further see data
tumor_vol_by_drug.boxplot(
    by='Drug Regimen',                    
    column=['Final Tumor Volume (mm3)'], 
    grid = False,
    rot=90,
    fontsize=10)
plt.title('Final Tumor Volume of Mouse\nper Drug Regimen', fontsize=18)
plt.suptitle('')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
show_treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Calculate the IQR and quantitatively determine if there are any potential outliers
def filter_outliers(values):
    # Create empty list to fill with tumor vol data (for plotting)
    filtered_values = []
    # get the quartile data
    q1 = np.quantile(values, 0.25)
    q3 = np.quantile(values, 0.75)
    iqr = q3 - q1
    #print(f'q1 {q1}, q3 {q3}, iqr {iqr}')
    for value in values:
        # Determine outliers using upper and lower bounds
        v_min = q1 - 1.5 * iqr
        v_max = q3 + 1.5 * iqr
        #print(f'value {value}, v_min {v_min}, v_max {v_max}')
        if (value < v_min) or (value > v_max):
            continue        
        else: # else, add the data
            filtered_values.append(value)
    # return filted values
    return filtered_values

# Filter out the outliers in the data sent in
def get_cleaned_tumor_vol_by_drug(data, remove_outliers=False, only_show=None):
    tumor_vol_by_drug = pd.DataFrame()
    for name, group in data.groupby('Drug Regimen')['Tumor Volume (mm3)']:
        # skip out the groups not in only_show
        if only_show and name not in only_show:
            continue
        tumor_vols = filter_outliers(group.values) if remove_outliers else group.values
        drug_vols = pd.DataFrame({'Drug Regimen': [name for n in tumor_vols],
                                  'Final Tumor Volume (mm3)':tumor_vols})
        tumor_vol_by_drug = pd.concat([tumor_vol_by_drug, drug_vols])
    return tumor_vol_by_drug.set_index('Drug Regimen')
    
tumor_vol_by_drug = get_cleaned_tumor_vol_by_drug(get_final_timepoints(mouse_study_data), 
                                                  remove_outliers=True,
                                                  only_show=show_treatments)

# display some info, just to see what's going on
tumor_vol_by_drug.groupby('Drug Regimen').describe()
```

```python
# Generate a box plot of the final tumor volume of each
# mouse across four regimens of interest

def get_treatments_data(tumor_vol_by_drug, treatments):
    data = []
    # add each treatment in same order
    for i in range(len(treatments)):
        tumor_vol = tumor_vol_by_drug.loc[treatments[i]]
        data.append(tumor_vol['Final Tumor Volume (mm3)'].values)
    return data

treatments_data = get_treatments_data(tumor_vol_by_drug,show_treatments)

fig = plt.figure() 
ax = fig.add_axes([0, 0, 1, 1]) 
bp = ax.boxplot(treatments_data) 
plt.title('Final Tumor Volume of Mouse\nper Drug Regimen', fontsize=18)
plt.xlabel('Drug Regimen')
plt.ylabel('Final Tumor Volume (mm3)')
plt.xticks(
    [n + 1 for n in range(len(show_treatments))], 
    show_treatments)
plt.show() 
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
drug_filter = 'Capomulin'

# get the tumor volume vs. time point for a mouse for a spesific drug
def get_tumor_vol_vs_timepoint(data, drug, mouse=float('inf')):
    # filter out nessisary data
    df = data[['Drug Regimen','Mouse ID','Timepoint','Tumor Volume (mm3)']]
    # filter out selected drug
    drug_df = df.loc[df['Drug Regimen'] == drug]
    # get all mice names
    mice = drug_df['Mouse ID'].unique()
    mouse_id = ''
    # ifs have value for mouse, use that value
    if (abs(mouse) >= len(mice)):
        mouse_id = random.choice(mice)
    else:
        # if anything goes wrong, set to default 0
        try:
            mouse_id = mice[mouse]
        except:
            mouse_id = mice[0]
    # return the corresponding mouse id
    return drug_df.loc[drug_df['Mouse ID'] == mouse_id]

tumor_vol_vs_timepoint = get_tumor_vol_vs_timepoint(
    mouse_study_data,
    'Capomulin')

x_values = tumor_vol_vs_timepoint['Timepoint']
y_values = tumor_vol_vs_timepoint['Tumor Volume (mm3)']
plt.plot(x_values, y_values)
plt.title('Tumor volume vs. time point for a\nmouse treated with Capomulin', fontsize=18)
plt.xlabel('Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

# send back the average tumor volume vs. mouse weight with the sent in
# drug name
def get_avg_tumor_vol_vs_mouse_weight(data, drug):
    df = data[['Drug Regimen','Mouse ID','Tumor Volume (mm3)','Weight (g)']]
    drug_df = df.loc[df['Drug Regimen'] == drug]
    return drug_df.groupby('Mouse ID').mean()

avg_tumor_vol_vs_mouse_weight = get_avg_tumor_vol_vs_mouse_weight(mouse_study_data,'Capomulin')
x_values = avg_tumor_vol_vs_mouse_weight['Weight (g)']
y_values = avg_tumor_vol_vs_mouse_weight['Tumor Volume (mm3)']
plt.scatter(x_values,y_values)
plt.title('Average tumor volume vs. mouse weight\nfor the Capomulin regimen', fontsize=18)
plt.xlabel('Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

avg_tumor_vol_vs_mouse_weight = get_avg_tumor_vol_vs_mouse_weight(mouse_study_data,'Capomulin')
x_values = avg_tumor_vol_vs_mouse_weight['Weight (g)']
y_values = avg_tumor_vol_vs_mouse_weight['Tumor Volume (mm3)']
(slope,intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
print()
regress_values = x_values * slope + intercept
line_eq = f'y = {round(slope, 2)}x + {round(intercept, 2)}'
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values, 'r-')
plt.annotate(line_eq, (20,38),fontsize=15,color='red')
plt.title('Average tumor volume vs. mouse weight\nfor the Capomulin regimen', fontsize=18)
plt.xlabel('Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
print(f'The r-squared(correlation coefficient) is {rvalue**2}')
```
